## Apply Perspective Transformation

Here you are going to apply a perspective transformation to get the birds-eye view of the lane lines. This is helpful because it cuts out a lot of unnecessary information in the image and makes it easier to fit a best-fit line to the lane lines. 

To apply this transformation you need to define four source points and four destination points. The source points define the region of interest which you would like to transform.  The four points must define a closed polygon. The destination points are the respective locations of the source points for the transformed perspective. 

It's generally a good idea to make your source points define a trapezoid in the lower half of the image, which captures the lane lines. Then define a square or rectange for the destination points. Play around with different combinations until you get a result you like. 

````python
# Draws viewing window onto copy of image
def draw_viewing_window(image,viewing_window):
    image_copy = np.copy(image)
    for line in viewing_window:
        for x1,y1,x2,y2 in line:
            cv2.line(image_copy,(x1,y1),(x2,y2),(0,255,0),5)
            
    return image_copy

# Makes the viewwing window which is used by draw_viewing_window()
def make_viewing_window(bottom_left,top_left,top_right,bottom_right):
    left_line = np.array([[bottom_left[0],bottom_left[1],top_left[0],top_left[1]]])
    top_line = np.array([[top_left[0],top_left[1],top_right[0],top_right[1]]])
    right_line = np.array([[top_right[0],top_right[1],bottom_right[0],bottom_right[1]]])
    bottom_line = np.array([[bottom_right[0],bottom_right[1],bottom_left[0],bottom_left[1]]])
    viewing_window = [left_line,top_line,right_line,bottom_line]
    
    return viewing_window

image_path = "test_images/test5.jpg"
image = cv2.imread(image_path)

# TODO: Fix the souce points
src_bottom_left = [100,780]
src_top_left = [125,300]
src_top_right = [550,300]
src_bottom_right = [690,480]
source_points = np.float32([src_bottom_left,src_top_left,src_top_right,
                 src_bottom_right])

# TODO: Fix the destination points
dest_bottom_left = [0,400]
dest_top_left = [1000,200]
dest_top_right = [10,70]
dest_bottom_right = [700,330]
dest_points = np.float32([dest_bottom_left,dest_top_left,dest_top_right,
                          dest_bottom_right])



M = cv2.getPerspectiveTransform(source_points,dest_points) # get the perspective matrix
img_size = (image.shape[1],image.shape[0])
birds_eye_image = cv2.warpPerspective(image, M, img_size,
                                      flags=cv2.INTER_LINEAR) # appply perspective transformation

src_viewing_window = make_viewing_window(src_bottom_left,src_top_left,src_top_right,src_bottom_right)
dest_viewing_window = make_viewing_window(dest_bottom_left,dest_top_left,dest_top_right,dest_bottom_right)

src_image_copy = draw_viewing_window(image,src_viewing_window) # draw source points
dest_image_copy = draw_viewing_window(birds_eye_image,dest_viewing_window) # draw destination points
 
# Display Results
f, axs = plt.subplots(1,2,figsize=(10,5))
axs = axs.ravel()
axs[0].imshow(src_image_copy)
axs[0].set_title('original')
axs[1].imshow(dest_image_copy)
axs[1].set_title('birds-eye view')
````

## Apply Color and Gradient Thresholdss



````python
# Applies sobel algorithm with respect to x or y axis with thresholded gradient
def abs_sobel_thresh(img, orient='x', sobel_kernel=7, thresh_min=0, thresh_max=255):
    
    
    # Convert to gray-scale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    # Take derivative
    if (orient == 'x'):
        # TODO: Take gradient with respect to x
        sobel = None
    else:
        # TODO: Take gradient with respect to y
        sobel = None
    
    # Take absolute value of derivative    
    abs_sobel = np.absolute(sobel)
    
    # Convert to 8-bit image (0 - 255)
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    # Make copy of scaled_sobel with all zeros
    sobel_binary = np.zeros_like(scaled_sobel)
    # Make all pixels within threshold range a value of 1
    # Keep all other pixels as 0
    sobel_binary[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
    binary_output = sobel_binary
    # binary_output = np.copy(img) # Remove this line
    return binary_output

# Function that applies Sobel x and y, 
# then computes the magnitude of the gradient
# and applies a threshold
def sobel_mag_thresh(img, sobel_kernel=3, mag_thresh=(0, 255)):

    thresh_min = mag_thresh[0]
    thresh_max = mag_thresh[1]
    
    # Convert to gray scale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # TODO: Take derivatives in both x and y direction
    sobelx = None
    sobely = None
    # TODO: Find magnitude of the gradient
    sobel_mag = None
    # Convert to 8-bit image (0 - 255)
    scaled_sobel = np.uint8(255*sobel_mag/np.max(sobel_mag))
    # Make a copy of sobel_mag with all zeros
    sobel_binary = np.zeros_like(scaled_sobel)
    # Make all pixels within threshold range a value of 1
    # Keep all other pixels as 0
    sobel_binary[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
    binary_output = sobel_binary

    return binary_output

# Function that applies Sobel x and y, 
# then computes the direction of the gradient
# and applies a threshold.
def dir_threshold(img, sobel_kernel=3, thresh=(0, np.pi/2)):
        
    # Min and Max Threshold Angles
    thresh_min = thresh[0]
    thresh_max = thresh[1]
    # Convert to gray scale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # TODO: Calculate the derivatives with respect to x and y
    sobelx = None
    sobely = None
    # Take absolute value of derivatives
    abs_sobelx = np.absolute(sobelx)
    abs_sobely = np.absolute(sobely)
    # Calculate angle for direction of gradient in radians using abs_sobelx and abs_sobely
    sobel_angle = None
    # Make a copy of sobel_angle with all zeros
    sobel_binary = np.zeros_like(sobel_angle)
    # Apply thresholding
    sobel_binary[(sobel_angle >= thresh_min) & (sobel_angle <= thresh_max)] = 1
    binary_output = sobel_binary
    
    return binary_output

image_path = "test_images/test5.jpg"
image = cv2.imread(image_path)

f, axs =plt.subplots(2, 2, figsize=(10, 10))
axs = axs.ravel()

gradx = abs_sobel_thresh(image, orient='x', sobel_kernel=7, thresh_min=40, thresh_max=100)

grady = abs_sobel_thresh(image, orient='y', sobel_kernel=7, thresh_min=40, thresh_max=100)

grad_mag = sobel_mag_thresh(image, sobel_kernel=7, mag_thresh=(40, 100))

grad_dir = dir_threshold(image, sobel_kernel=3, thresh=(0.7, 1.3))

axs[0].imshow(gradx)
axs[0].set_title('gradx')
axs[1].imshow(grady)
axs[1].set_title('grady')
axs[2].imshow(grad_mag)
axs[2].set_title('grad_mag')
axs[3].imshow(grad_dir)
axs[3].set_title('grad_dir')

plt.show()
````
